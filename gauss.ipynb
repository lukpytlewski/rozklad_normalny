```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.DataFrame(np.random.normal(3,1,100))
df2 = pd.DataFrame(np.random.normal(3,1,10000))
```

```python
df1.columns = ['value']
df1.head()
```

```python
df2.columns = ['value']
df2.head()
```

```python
stats1 = df1[['value']].describe(percentiles=[.25, .5, .75, 1])
stats1.loc['IQR'] = stats1.loc['75%'] - stats1.loc['25%']
stats1
```

```python
stats2 = df2[['value']].describe(percentiles=[.25, .5, .75, 1])
stats2.loc['IQR'] = stats2.loc['75%'] - stats2.loc['25%']
stats2
```

```python
df3=pd.concat([stats1, stats2], axis=1)
df3
```

```python
df3.columns=['statystyki opisowe dla sample 100', 'statystyki opisowe dla sample 10000']
df3
```

```python
plt.hist(df1['value'], label = 'sample 100', density=True, edgecolor='black', bins=30, alpha=0.5)
plt.hist(df2['value'], label = 'sample = 10 000', density=True, edgecolor='red', bins=30, alpha=0.5)
plt.legend()
plt.show()
```

Próba 10 tys. pozwala bardziej precyzyjnie oszacować statystyki opisowe dla określonego rozkładu normalnego. Średnia dla próby 10 tys. jest bliższa wartości oczekiwanej niż w przypadku średniej dla próby 100. Podobnie sprawa ma się z oszacowaniem odchylenia standardowe, które jest precyzyjniej oszacowane dla próby liczącej 10 tys.
